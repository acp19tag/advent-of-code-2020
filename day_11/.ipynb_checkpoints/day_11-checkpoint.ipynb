By modeling the process people use to choose (or abandon) their seat in the waiting area, you're pretty sure you can predict the best place to sit. You make a quick map of the seat layout (your puzzle input).

Now, you just need to model the people who will be arriving shortly. Fortunately, people are entirely predictable and always follow a simple set of rules. All decisions are based on the number of occupied seats adjacent to a given seat (one of the eight positions immediately up, down, left, right, or diagonal from the seat). The following rules are applied to every seat simultaneously:

    If a seat is empty (L) and there are no occupied seats adjacent to it, the seat becomes occupied.
    If a seat is occupied (#) and four or more seats adjacent to it are also occupied, the seat becomes empty.
    Otherwise, the seat's state does not change.

Floor (.) never changes; seats don't move, and nobody sits on the floor.

```python
# because ALL DECISIONS ARE BASED ON THE NUMBER OF OCCUPIED, ADJACENT SEATS, 
# we capture this info

def count_adjacent_occupied_seats(row_index, col_index, seating_grid):
    
    # are there rows above?
    if row_index > 0:
        rows_above = True
    else:
        rows_above = False
        
    # are there rows below?
    if row_index < len(seating_grid)-1:
        rows_below = True
    else:
        rows_below = False
        
    # are there columns to the left? 
    if col_index > 0: 
        cols_left = True
    else: 
        cols_left = False
        
    # are there columns to the right? 
    if col_index < len(seating_grid[0])-1:
        cols_right = True
    else:
        cols_right = False
        
    # counting occupied seats...
    counter = 0
    
    # northwest
    if rows_above and cols_left:
        if seating_grid[row_index-1][col_index-1] == '#':
            counter +=1
    
    # north
    if rows_above:
        if seating_grid[row_index-1][col_index] == '#':
            counter +=1
    
    # northeast
    if rows_above and cols_right:
        if seating_grid[row_index-1][col_index+1] == '#':
            counter +=1
            
    # west
    if cols_left:
        if seating_grid[row_index][col_index-1] == '#':
            counter +=1
            
    # east
    if cols_right:
        if seating_grid[row_index][col_index+1] == '#':
            counter +=1
            
    # southwest
    if rows_below and cols_left:
        if seating_grid[row_index+1][col_index-1] == '#':
            counter +=1
            
    # south
    if rows_below:
        if seating_grid[row_index+1][col_index] == '#':
            counter +=1
            
    # southeast
    if rows_below and cols_right:
        if seating_grid[row_index+1][col_index+1] == '#':
            counter +=1
    
    return counter
    
```

```python
# create dict in format {(row_index, col_index): {'position': position, adjacent_occupied: count}}

def seating_grid_to_dict(seating_grid):
    
    seating_grid_dict = {}
    
    for row_index, row_layout in enumerate(seating_grid):
        
        for col_index, position in enumerate(row_layout):
            
            seating_grid_dict[(row_index, col_index)] = {
                'position': position,
                'adjacent_occupied': count_adjacent_occupied_seats(row_index, col_index, seating_grid)
            }
            
    return seating_grid_dict
```

```python
# and going back again

def seating_dict_to_grid(seating_dict):
    
    seating_grid = []
    max_row = max([x[0] for x in seating_dict.keys()])
    max_col = max([x[1] for x in seating_dict.keys()])
    
    for row_index in range(max_row+1):
        
        row = ''
        
        for col_index in range(max_col+1):
            
            row += seating_dict[(row_index, col_index)]['position']
            
            if col_index == max_col:
                
                seating_grid.append(row)
                
    return seating_grid
    
```

```python
def update_seating_grid(seating_grid):
    
    seating_dict = seating_grid_to_dict(seating_grid)
    
    for grid_space in seating_dict:
        
        # if a seat is empty and there are no occupied seats, it becomes occupied
        if seating_dict[grid_space]['position'] == 'L' and seating_dict[grid_space]['adjacent_occupied'] == 0:
            
            seating_dict[grid_space] = {'position': '#'}
        
        # if a seat is occupied and four or more adjacent seats are occupied, it becomes empty
        elif seating_dict[grid_space]['position'] == '#' and seating_dict[grid_space]['adjacent_occupied'] >= 4:
            
            seating_dict[grid_space] = {'position': 'L'}
            
    return seating_dict_to_grid(seating_dict)
```

```python
# function to count occupied seats
def count_occupied_seats(seating_grid):
    
    count_by_row = [row.count('#') for row in seating_grid]
    
    return sum(count_by_row)
```

```python
def find_static_seat_positions(seating_grid):
    
    current_position = seating_grid
    previous_position = []
    iteration_count = 0
    
    while current_position != previous_position:
        
        previous_position = current_position
        
        current_position = update_seating_grid(current_position)
        
        iteration_count += 1
        
    return count_occupied_seats(current_position), iteration_count
```

```python
# we need a re-write of the adjacent seat counter

def count_visible_occupied_seats(row_index, col_index, seating_grid):
    
    max_row = len(seating_grid)-1
    max_col = len(seating_grid[0])-1
    
    # are there rows above?
    if row_index > 0:
        rows_above = True
    else:
        rows_above = False
        
    # are there rows below?
    if row_index < max_row:
        rows_below = True
    else:
        rows_below = False
        
    # are there columns to the left? 
    if col_index > 0: 
        cols_left = True
    else: 
        cols_left = False
        
    # are there columns to the right? 
    if col_index < max_col:
        cols_right = True
    else:
        cols_right = False
        
    # counting occupied seats...
    counter = 0
    
    # northwest
    if rows_above and cols_left:
        
        seat_seen = False
        
        viewing_row_index = row_index
        viewing_col_index = col_index
        
        while not seat_seen:
            
            viewing_row_index -= 1
            viewing_col_index -= 1
            
            if not (0 <= viewing_row_index <= max_row and 0 <= viewing_col_index <= max_col):
                
                break
                
            viewing_position = seating_grid[viewing_row_index][viewing_col_index]

            if viewing_position != '.':
                
                seat_seen = True
                
                if viewing_position == '#':
                    
                    counter += 1
    
    # north
    if rows_above:

        seat_seen = False
        
        viewing_row_index = row_index
        viewing_col_index = col_index
        
        while not seat_seen:
            
            viewing_row_index -= 1
            
            if not (0 <= viewing_row_index <= max_row and 0 <= viewing_col_index <= max_col):
                
                break
                
            viewing_position = seating_grid[viewing_row_index][viewing_col_index]

            if viewing_position != '.':
                
                seat_seen = True
                
                if viewing_position == '#':
                    
                    counter += 1
    
    # northeast
    if rows_above and cols_right:
        
        seat_seen = False
        
        viewing_row_index = row_index
        viewing_col_index = col_index
        
        while not seat_seen:
        
            viewing_row_index -= 1
            viewing_col_index += 1
            
            if not (0 <= viewing_row_index <= max_row and 0 <= viewing_col_index <= max_col):
                
                break
                
            viewing_position = seating_grid[viewing_row_index][viewing_col_index]

            if viewing_position != '.':
                
                seat_seen = True
                
                if viewing_position == '#':
                    
                    counter += 1
            
    # west
    if cols_left:
       
        seat_seen = False
        
        viewing_row_index = row_index
        viewing_col_index = col_index
        
        while not seat_seen:
        
            viewing_col_index -= 1
            
            if not (0 <= viewing_row_index <= max_row and 0 <= viewing_col_index <= max_col):
                
                break
                
            viewing_position = seating_grid[viewing_row_index][viewing_col_index]

            if viewing_position != '.':
                
                seat_seen = True
                
                if viewing_position == '#':
                    
                    counter += 1
            
    # east
    if cols_right:

        seat_seen = False
        
        viewing_row_index = row_index
        viewing_col_index = col_index
        
        while not seat_seen:
            
            viewing_col_index += 1
            
            if not (0 <= viewing_row_index <= max_row and 0 <= viewing_col_index <= max_col):
                
                break
                
            viewing_position = seating_grid[viewing_row_index][viewing_col_index]

            if viewing_position != '.':
                
                seat_seen = True
                
                if viewing_position == '#':
                    
                    counter += 1
            
    # southwest
    if rows_below and cols_left:

        seat_seen = False
        
        viewing_row_index = row_index
        viewing_col_index = col_index
        
        while not seat_seen:
        
            viewing_row_index += 1
            viewing_col_index -= 1
            
            if not (0 <= viewing_row_index <= max_row and 0 <= viewing_col_index <= max_col):
                
                break
                
            viewing_position = seating_grid[viewing_row_index][viewing_col_index]

            if viewing_position != '.':
                
                seat_seen = True
                
                if viewing_position == '#':
                    
                    counter += 1
            
    # south
    if rows_below:

        seat_seen = False
        
        viewing_row_index = row_index
        viewing_col_index = col_index
        
        while not seat_seen:
        
            viewing_row_index += 1
            
            if not (0 <= viewing_row_index <= max_row and 0 <= viewing_col_index <= max_col):
                
                break
                
            viewing_position = seating_grid[viewing_row_index][viewing_col_index]

            if viewing_position != '.':
                
                seat_seen = True
                
                if viewing_position == '#':
                    
                    counter += 1
            
    # southeast
    if rows_below and cols_right:
       
        seat_seen = False
        
        viewing_row_index = row_index
        viewing_col_index = col_index
        
        while not seat_seen:
        
            viewing_row_index += 1
            viewing_col_index += 1
            
            if not (0 <= viewing_row_index <= max_row and 0 <= viewing_col_index <= max_col):
                
                break
                
            viewing_position = seating_grid[viewing_row_index][viewing_col_index]

            if viewing_position != '.':
                
                seat_seen = True
                
                if viewing_position == '#':
                    
                    counter += 1
    
    return counter
```

```python
# update function to include the new visible seat rule

def seating_grid_to_dict(seating_grid):
    
    seating_grid_dict = {}
    
    for row_index, row_layout in enumerate(seating_grid):
        
        for col_index, position in enumerate(row_layout):
            
            seating_grid_dict[(row_index, col_index)] = {
                'position': position,
                'adjacent_occupied': count_visible_occupied_seats(row_index, col_index, seating_grid)
            }
            
    return seating_grid_dict
```

```python
# update function to take '5 or more seat' rule into account

def update_seating_grid(seating_grid):
    
    seating_dict = seating_grid_to_dict(seating_grid)
    
    for grid_space in seating_dict:
        
        # if a seat is empty and there are no occupied seats, it becomes occupied
        if seating_dict[grid_space]['position'] == 'L' and seating_dict[grid_space]['adjacent_occupied'] == 0:
            
            seating_dict[grid_space] = {'position': '#'}
        
        # if a seat is occupied and four or more adjacent seats are occupied, it becomes empty
        elif seating_dict[grid_space]['position'] == '#' and seating_dict[grid_space]['adjacent_occupied'] >= 5:
            
            seating_dict[grid_space] = {'position': 'L'}
            
    return seating_dict_to_grid(seating_dict)
```
