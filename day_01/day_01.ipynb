```python
def expense_checkr_2(puzzle_input):

    no_match = True

    for number in puzzle_input:

        new_goal = 2020 - number

        for second_number in puzzle_input:

            if new_goal - second_number in puzzle_input and no_match:

                answer_1, answer_2, answer_3 = number, second_number, new_goal - second_number
                no_match = False

    if no_match:
        return 'NO MATCH FOUND'
    else:
        return answer_1 * answer_2 * answer_3
```
